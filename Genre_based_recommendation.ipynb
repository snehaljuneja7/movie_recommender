```python
#Genre based movie recommendation system 
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')

movies.head()
```

```python
movies.columns
```

```python
#Dropping columns that are not useful
features = ['genres','popularity','vote_average','id','title', 'vote_count']
movies = movies[features]
```

```python
movies.head()
```

```python
#Extract genre names from object
from ast import literal_eval

def convert(obj):
    list = []
    for j in literal_eval(obj):
        list.append(j['name'])
    return list
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres'].iloc[0]
```

```python
#Convert genres list to | seperated string
def convert2(obj):
    temp = ""
    temp = "|".join(str(e) for e in obj)
    return temp
```

```python
convert2(['Action', 'Adventure', 'Fantasy', 'Science Fiction'])
```

```python
movies['genres'] = movies['genres'].apply(convert2)
```

```python
movies.head()
```

```python
#Get list of unique genres
genre_list = ""
for index,row in movies.iterrows():
        genre_list += row.genres + "|"
#Split the string into a list of values
genre_list_split = genre_list.split('|')
#De-duplicate values
new_list = list(set(genre_list_split))
#Remove the value that is blank
new_list.remove('')
#Inspect list of genres
new_list
```

```python
#Add new column for each genre in dataframe

movies_genres = movies.copy()
for genre in new_list :
    movies_genres[genre] = movies_genres.apply(lambda _:int(genre in _.genres), axis = 1)
```

```python
movies_genres.head()
```

```python
#calculate the percentile count. It gives the no of ratings at least 70% of the movies have
np.percentile(movies_genres['vote_count'],70)
```

```python
#Get the average movie rating across all movies 
avg_rating_all=movies_genres['vote_average'].mean()
avg_rating_all
```

```python
#Function for weighted rating score based off count of votes
def weighted_rating(x, m=10, C=avg_rating_all):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
#Calculate the weighted score for each movie
movies_genres['weighted_score'] = movies_genres.apply(weighted_rating, axis=1)
movies_genres.head()
```

```python
#list top scored movies over the whole range of movies
pd.DataFrame(movies_genres.sort_values(['weighted_score'],ascending=False)[['title','vote_count','vote_average','weighted_score','genres']][:10])
```

```python
# Gives the best movies according to genre based on weighted score which is calculated using IMDB formula
def get_genre_based_recommendation(genre,top_n):
    temp = movies_genres.loc[(movies_genres[genre]==1)]
    temp = temp.sort_values(by=['weighted_score'],ascending=False)
    temp = temp.iloc[:top_n]
    features = ['title','genres','weighted_score']
    temp = temp[features]
    return temp
```

```python
temp = get_genre_based_recommendation('Drama',5)
temp.head()
```
